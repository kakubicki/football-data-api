```python
import pandas as pd 
import numpy as np
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import re
```

```python
url = 'https://fbref.com/en/comps/12/La-Liga-Stats'
response = requests.get(url)
soup = BeautifulSoup(response.text,'html.parser')

tables = soup.find_all('table', attrs = {'class': 'stats_table'})
```

```python
table_overall = tables[2]
rows = table_overall.find_all('tr')
```

```python
headers = [th.text.strip() for th in rows[1].find_all('th')]
```

```python
data = []
for row in rows[2:]:
    index_value = row.find('th').text.strip()
    row_data = [index_value] + [td.text.strip() for td in row.find_all('td')]
    data.append(row_data)
```

```python
df = pd.DataFrame(data, columns=headers)
```
